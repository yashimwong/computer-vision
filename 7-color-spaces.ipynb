```python
import cv2
import matplotlib.pyplot as plot
%matplotlib inline
```

```python
img = cv2.imread("dog_backpack.jpg")
plot.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plot.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
plot.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
plot.imshow(img)
```
