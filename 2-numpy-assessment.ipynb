### Task: Use PIL and matplotlib to read and display test2.jpg

```python
import matplotlib.pyplot as plot
from PIL import Image
```

```python
img = Image.open('./test2.jpg')
plot.imshow(img)
```

```python
img_array = np.asarray(img)
img_array
```

## Final Task: Use slicing to set the Red and Green channels of the picture to 0, then use imshow() to show the isolated blue channel

```python
img_blue = img_array.copy()
```

```python
# Remove Red channel
img_blue[:, :, 0] = 0

# Remove Green color
img_blue[:, :, 1] = 0
```

```python
plot.imshow(img_blue)
```
