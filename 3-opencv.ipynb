## OpenCV Image Reading and Color Spaces

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
img = cv2.imread('./test.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
# Matplotlib --> Red Green Blue (RGB)
# OpenCV --> Blue Green Red (BGR)
new_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(new_img)
```

```python
img_gray = cv2.imread('./test.jpg', cv2.IMREAD_GRAYSCALE)
img_gray.shape
```

```python
plt.imshow(img_gray, cmap="gray")
```

```python
new_resized_img = cv2.resize(new_img,(1000, 400))
plt.imshow(new_resized_img)
```

```python
# Resize according to ratio (50%)
new_ratio_resized_img = cv2.resize(new_img, (0, 0), new_img, 0.5, 0.5)
plt.imshow(new_ratio_resized_img)
```

## Flipping Images

```python
flipped_new_img = cv2.flip(new_img, 0)
plt.imshow(flipped_new_img)
```

```python
flipped_new_img = cv2.flip(new_img, 1)
plt.imshow(flipped_new_img)
```

```python
flipped_new_img = cv2.flip(new_img, -1)
plt.imshow(flipped_new_img)
```

## Save Image File

```python
cv2.imwrite('./new.jpg', img_gray)
```
