### Task: Open the dog_backpack.jpg image and display it in the notebook. Make sure to correct for the RGB Order

```python
import cv2
import numpy as np
import matplotlib.pyplot as plot

img = cv2.imread("dog_backpack.jpg")
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plot.imshow(img)
```

### Task: Flip the image upside down and display it in the notebook.

```python
flipped_img = cv2.flip(img, 0)
plot.imshow(flipped_img)
```

```python
plot.imshow(img)
```

### Task: Draw an empty Red rectangle around the dogs face and display the image in the notebook

```python
img1 = img.copy()
cv2.rectangle(img1, pt1=(150, 350), pt2=(650, 750), color=(255, 0, 0), thickness=14)
plot.imshow(img1)
```

### Task: Draw a blue triangle in the middle of the image.

```python
img2 = img.copy()
cv2.line(img2, pt1=(150,780), pt2=(500,400), color=(0, 0, 255), thickness=14)
cv2.line(img2, pt1=(500,400), pt2=(850,780), color=(0, 0, 255), thickness=14)
cv2.line(img2, pt1=(850,780), pt2=(150,780), color=(0, 0, 255), thickness=14)
plot.imshow(img2)
```

### Task: fill in the drawn triangle

```python
img3 = img.copy()
contours = np.array([[150, 780], [500, 400], [850, 780]])
cv2.fillPoly(img3, pts=[contours], color=(0, 0, 255))
plot.imshow(img3)
```
